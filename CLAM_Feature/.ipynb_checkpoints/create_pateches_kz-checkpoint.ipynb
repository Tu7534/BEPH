```python
import os
import h5py
import numpy as np
import pandas as pd
import scanpy as sc
from PIL import Image
import matplotlib.pyplot as plt

# ================= 配置区域 =================
h5ad_path = 'GSM6177623.h5ad'
image_path = 'GSM6177623_tissue_hires_image.png'
save_dir = './visium_results_resized'

# --- 关键修改 ---
src_size = 48      # 1. 物理裁剪大小：在原图上切多少像素 (建议 64 或 96)
target_size = 224  # 2. 目标大小：放大到多少像素喂给模型 (BEPH/CLAM 需要 224)
# ----------------

save_patches_as_images = True 

def process_visium_data_resized(h5ad_path, image_path, save_dir, src_size=64, target_size=224, save_imgs=False):
    slide_id = os.path.splitext(os.path.basename(image_path))[0]
    patch_save_dir = os.path.join(save_dir, 'patches')
    img_save_dir = os.path.join(patch_save_dir, f'{slide_id}_imgs')
    os.makedirs(img_save_dir, exist_ok=True)
    
    h5_save_path = os.path.join(patch_save_dir, f'{slide_id}.h5')
    
    print(f"正在处理: {slide_id}")
    print(f"策略: 从原图切 {src_size}x{src_size} -> 放大至 {target_size}x{target_size}")

    # 1. 读取数据
    try:
        adata = sc.read_h5ad(h5ad_path)
        Image.MAX_IMAGE_PIXELS = None 
        full_image = Image.open(image_path).convert('RGB')
    except Exception as e:
        print(f"读取失败: {e}")
        return

    # 2. 获取缩放因子与坐标
    coords = adata.obsm['spatial']
    scale_factor = 1.0
    library_id = list(adata.uns.get('spatial', {}).keys())[0] if adata.uns.get('spatial') else None
    
    if library_id:
        scalefactors = adata.uns['spatial'][library_id].get('scalefactors', {})
        if 'hires' in image_path:
            scale_factor = scalefactors.get('tissue_hires_scalef', 1.0)
        elif 'lowres' in image_path:
            scale_factor = scalefactors.get('tissue_lowres_scalef', 1.0)
            
    # 计算中心点像素坐标
    pixel_coords = (coords * scale_factor).astype(int)

    # 3. 计算左上角坐标 (基于 src_size)
    # 注意：BEPH 的 h5 文件里存的坐标应该是基于"哪张图"的？
    # 通常 BEPH/CLAM 的流程是：读取 h5 坐标 -> 回到 WSI (这里是 hires) 读取 -> 裁剪 -> 实时 Resize
    # 因此，我们这里保存的坐标，必须是针对【原图切取位置】的坐标。
    half_size = src_size // 2
    top_left_coords = pixel_coords - half_size
    
    img_w, img_h = full_image.size
    valid_coords = []
    valid_indices = []

    for idx, (x, y) in enumerate(top_left_coords):
        if x >= 0 and y >= 0 and x + src_size <= img_w and y + src_size <= img_h:
            valid_coords.append([x, y])
            valid_indices.append(idx)
            
    valid_coords = np.array(valid_coords)
    print(f"有效 Patch 数: {len(valid_coords)}")

    # 4. 保存坐标到 .h5
    # 重要：这里保存的 patch_size 属性应该是 src_size (64)，因为这是去原图切的大小
    # 但 patch_level 设为 0
    # 为了适配某些特定 pipeline，有时需要在 custom_transforms 里做 resize
    with h5py.File(h5_save_path, 'w') as f:
        dset = f.create_dataset('coords', data=valid_coords)
        dset.attrs['patch_size'] = src_size  # 记录原始切图大小
        dset.attrs['patch_level'] = 0
        f.attrs['n_patches'] = len(valid_coords)
    print(f"坐标已保存至: {h5_save_path}")

    # 5. 物理裁剪并放大保存 (用于可视化或直接训练)
    if save_imgs:
        print(f"正在裁剪并放大保存图片...")
        for i, (x, y) in enumerate(valid_coords):
            if i % 100 == 0: print(f"进度: {i}/{len(valid_coords)}", end='\r')
            
            # (1) 裁剪原始大小 (e.g. 64x64)
            patch = full_image.crop((x, y, x + src_size, y + src_size))
            
            # (2) 放大到目标大小 (e.g. 224x224)
            # 使用 BICUBIC 或 LANCZOS 高质量插值
            patch_resized = patch.resize((target_size, target_size), Image.Resampling.LANCZOS)
            
            patch_resized.save(os.path.join(img_save_dir, f"patch_{i}_{x}_{y}.png"))
            
        print(f"\n全部完成。图片保存在: {img_save_dir}")

# 执行
process_visium_data_resized(h5ad_path, image_path, save_dir, src_size, target_size, save_patches_as_images)
```

## 处理全部数据

```python
import os
import h5py
import numpy as np
import pandas as pd
import scanpy as sc
from PIL import Image
from tqdm import tqdm  # 进度条库，如果没有请 pip install tqdm

# ================= 1. 全局配置路径 (请确认这里) =================
# 项目根目录
ROOT_DIR = './kz_data' 

# 输入目录
H5AD_DIR = os.path.join(ROOT_DIR, 'Raw_Data', 'h5ad_files')
IMAGE_DIR = os.path.join(ROOT_DIR, 'Raw_Data', 'images')
CSV_PATH = 'process_list.csv'  # 你的索引文件

# 输出目录
OUTPUT_DIR = os.path.join(ROOT_DIR, 'Segmentation')
PATCHES_SAVE_DIR = os.path.join(OUTPUT_DIR, 'patches')  # 存放 .h5 坐标
MASKS_SAVE_DIR = os.path.join(OUTPUT_DIR, 'masks')      # (可选) 存放可视化图

# 参数设置
SRC_SIZE = 48       # 在原图上切的大小
TARGET_SIZE = 224   # 喂给模型的大小
SAVE_PATCH_IMGS = False # ⚠ 建议设为 False，除非你需要肉眼检查。后续特征提取不需要这些小图。

# ================= 2. 核心处理函数 =================
def process_single_sample(slide_id, h5ad_path, image_path):
    """处理单个样本的核心逻辑"""
    
    # 检查文件是否存在
    if not os.path.exists(h5ad_path):
        print(f"❌ [跳过] 找不到 h5ad 文件: {h5ad_path}")
        return False
    if not os.path.exists(image_path):
        print(f"❌ [跳过] 找不到图片文件: {image_path}")
        return False

    try:
        # 1. 读取数据
        adata = sc.read_h5ad(h5ad_path)
        
        # 允许加载超大图片
        Image.MAX_IMAGE_PIXELS = None 
        # 为了获取尺寸，我们这里只open不load，节省内存。如果是存图模式再load
        full_image = Image.open(image_path) 
        img_w, img_h = full_image.size

        # 2. 获取缩放因子与坐标
        if 'spatial' not in adata.obsm.keys():
            print(f"⚠️ {slide_id}: .h5ad 中没有 spatial 坐标，跳过。")
            return False

        coords = adata.obsm['spatial']
        scale_factor = 1.0
        
        # 尝试自动获取缩放因子
        library_id = list(adata.uns.get('spatial', {}).keys())[0] if adata.uns.get('spatial') else None
        if library_id:
            scalefactors = adata.uns['spatial'][library_id].get('scalefactors', {})
            # 根据文件名判断是用 hires 还是 lowres 因子 (你的文件名包含 hires)
            if 'hires' in os.path.basename(image_path):
                scale_factor = scalefactors.get('tissue_hires_scalef', 1.0)
            elif 'lowres' in os.path.basename(image_path):
                scale_factor = scalefactors.get('tissue_lowres_scalef', 1.0)
        
        # 计算中心点像素坐标
        pixel_coords = (coords * scale_factor).astype(int)

        # 3. 计算左上角坐标 (Top-Left)
        # BEPH/CLAM 需要的是 Patch 左上角的坐标
        half_size = SRC_SIZE // 2
        top_left_coords = pixel_coords - half_size
        
        valid_coords = []
        
        # 边界检查
        for x, y in top_left_coords:
            if x >= 0 and y >= 0 and x + SRC_SIZE <= img_w and y + SRC_SIZE <= img_h:
                valid_coords.append([x, y])
        
        valid_coords = np.array(valid_coords)

        # 4. 保存坐标 .h5 (这是后续特征提取必须的文件)
        h5_save_path = os.path.join(PATCHES_SAVE_DIR, f'{slide_id}.h5')
        
        with h5py.File(h5_save_path, 'w') as f:
            # 数据集名称必须叫 'coords'，这是 CLAM 的规定
            dset = f.create_dataset('coords', data=valid_coords)
            dset.attrs['patch_size'] = SRC_SIZE 
            dset.attrs['patch_level'] = 0
            f.attrs['n_patches'] = len(valid_coords)
        
        # 5. (可选) 保存裁剪后的小图片
        if SAVE_PATCH_IMGS:
            img_save_subdir = os.path.join(OUTPUT_DIR, 'patch_images', f'{slide_id}')
            os.makedirs(img_save_subdir, exist_ok=True)
            full_image = full_image.convert('RGB') # 只有要切图时才转换
            
            for i, (x, y) in enumerate(valid_coords):
                patch = full_image.crop((x, y, x + SRC_SIZE, y + SRC_SIZE))
                patch_resized = patch.resize((TARGET_SIZE, TARGET_SIZE), Image.Resampling.LANCZOS)
                patch_resized.save(os.path.join(img_save_subdir, f"p_{i}_{x}_{y}.png"))

        return True

    except Exception as e:
        print(f"❌ 处理 {slide_id} 时发生错误: {e}")
        return False

# ================= 3. 主程序入口 =================
if __name__ == '__main__':
    # 确保输出目录存在
    os.makedirs(PATCHES_SAVE_DIR, exist_ok=True)
    
    # 1. 读取 CSV 列表
    if not os.path.exists(CSV_PATH):
        print(f"错误：找不到索引文件 {CSV_PATH}")
        exit()
        
    df = pd.read_csv(CSV_PATH)
    # 假设第一列是 ID，如果你的列名不是 'slide_id'，请修改这里
    # 例如：ids = df['sample_name'].tolist()
    if 'slide_id' in df.columns:
        ids = df['slide_id'].tolist()
    else:
        ids = df.iloc[:, 0].tolist() # 默认取第一列
    
    print(f"📋 准备处理 {len(ids)} 个样本...")
    print(f"📂 输入: {H5AD_DIR}")
    print(f"📂 输出: {PATCHES_SAVE_DIR}")

    # 2. 批量循环
    success_count = 0
    # 使用 tqdm 显示进度条
    for slide_id in tqdm(ids, desc="Processing"):
        # 构造完整路径
        # 注意：这里假设 h5ad 和 png 的文件名（不含后缀）与 CSV 中的 ID 完全一致
        h5ad_file = os.path.join(H5AD_DIR, f"{slide_id}.h5ad")
        
        # 寻找对应的图片文件 (防止 png 后缀大小写问题，或者有些是 tissue_hires_image.png)
        # 如果你的图片名严格是 {ID}.png，直接用下面这行：
        img_file = os.path.join(IMAGE_DIR, f"{slide_id}.png")
        
        # 如果图片名可能带后缀 (如 ID_tissue_hires_image.png)，需要自己调整这里
        # img_file = os.path.join(IMAGE_DIR, f"{slide_id}_tissue_hires_image.png")

        if process_single_sample(str(slide_id), h5ad_file, img_file):
            success_count += 1
            
    print(f"\n✨ 处理完成！成功: {success_count} / 总数: {len(ids)}")
    print(f"坐标文件已保存在: {PATCHES_SAVE_DIR}")
```
