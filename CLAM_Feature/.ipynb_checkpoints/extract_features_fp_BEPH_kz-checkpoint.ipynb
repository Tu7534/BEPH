```python
import os
import time
import h5py
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from mmengine.config import Config
from mmengine.registry import init_default_scope
from mmselfsup.apis import init_model

# ================= 配置区域 (请修改这里) =================
# 1. 输入: 上一步生成的 h5 文件路径 (注意：这里直接指定具体的 .h5 文件)
h5_path = '/data/home/wangzz_group/zhaipengyuan/BEPH-main/DATA_DIRECTORY/kz/visium_test_data/visium_results_resized/patches/GSM6177623_tissue_hires_image.h5'

# 2. 输入: 原始 Visium 图片路径
slide_path = 'GSM6177623_tissue_hires_image.png'

# 3. 输出: 特征保存目录
output_dir = '/data/home/wangzz_group/zhaipengyuan/BEPH-main/DATA_DIRECTORY/kz/visium_features/'

# 4. 模型配置 (请修改为您本地 BEPH 仓库的实际路径)
# 注意：beitv2_base.py 是 mmselfsup 的配置文件
model_config = '/data/home/wangzz_group/zhaipengyuan/BEPH-main/mmselfsup/configs/tsne/beitv2_base.py' 
model_checkpoint = '/data/home/wangzz_group/zhaipengyuan/BEPH-main/checkpoints/BEPH_backbone.pth' # 模型权重文件

# 5. 参数设置
target_patch_size = 224  # 模型需要的输入大小 (不要改)
batch_size = 48          # 显存够大可以改大，比如 64 或 128
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
# =========================================================

# --- 辅助类：模拟 OpenSlide，让代码能读取普通 PNG ---
class PILSlide:
    def __init__(self, path):
        Image.MAX_IMAGE_PIXELS = None
        self.img = Image.open(path).convert('RGB')
        # 模拟 openslide 的 level_dimensions 属性，只有一层 level 0
        self.level_dimensions = [self.img.size] 
        
    def read_region(self, location, level, size):
        # location 是 (x, y) 左上角坐标
        # level 在这里被忽略，因为只有一层
        # size 是 (w, h)
        x, y = location
        w, h = size
        # 裁剪并返回图像
        return self.img.crop((x, y, x + w, y + h))

# --- Dataset 定义 (精简版 Whole_Slide_Bag_FP) ---
def eval_transforms(mean, std):
    return transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=mean, std=std)
    ])

class Whole_Slide_Bag_FP(Dataset):
    def __init__(self, file_path, wsi, target_patch_size=224):
        self.wsi = wsi
        self.file_path = file_path
        self.target_patch_size = (target_patch_size, target_patch_size)
        
        # BEPH/ImageNet 归一化参数
        self.roi_transforms = eval_transforms(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))

        with h5py.File(self.file_path, "r") as f:
            dset = f['coords']
            # 从 h5 属性中读取生成时保存的 patch_size (比如 48)
            self.patch_size = f['coords'].attrs['patch_size']
            self.patch_level = f['coords'].attrs['patch_level']
            self.length = len(dset)
            
        print(f"Dataset initialized: {self.length} patches.")
        print(f"Source Patch Size: {self.patch_size} -> Target Resize: {self.target_patch_size}")

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        with h5py.File(self.file_path, 'r') as hdf5_file:
            coord = hdf5_file['coords'][idx]
        
        # 从大图中读取原始大小 (例如 48x48)
        img = self.wsi.read_region(coord, self.patch_level, (self.patch_size, self.patch_size)).convert('RGB')

        # 实时缩放到模型需要的 224x224
        if self.target_patch_size is not None:
            img = img.resize(self.target_patch_size, Image.Resampling.LANCZOS)
            
        img = self.roi_transforms(img) # 注意：移除了 unsqueeze(0)，DataLoader 会自动堆叠
        return img, coord

def collate_features(batch):
    img = torch.stack([item[0] for item in batch], dim=0)
    coords = np.vstack([item[1] for item in batch])
    return [img, coords]

# --- 主执行逻辑 ---
def run_feature_extraction():
    # 1. 准备目录
    os.makedirs(os.path.join(output_dir, 'h5_files'), exist_ok=True)
    os.makedirs(os.path.join(output_dir, 'pt_files'), exist_ok=True)
    
    slide_id = os.path.splitext(os.path.basename(h5_path))[0]
    output_h5_name = os.path.join(output_dir, 'h5_files', slide_id + '.h5')
    output_pt_name = os.path.join(output_dir, 'pt_files', slide_id + '.pt')
    
    print(f"开始提取特征: {slide_id}")
    
    # 2. 初始化模型
    print("Loading model...")
    try:
        cfg = Config.fromfile(model_config)
        init_default_scope(cfg.get('default_scope', 'mmselfsup'))
        model = init_model(cfg, model_checkpoint, device=device)
        model.eval()
        if torch.cuda.device_count() > 1:
            model = nn.DataParallel(model)
        print("Model loaded successfully.")
    except Exception as e:
        print(f"Error loading model: {e}")
        print("请检查 model_config 和 model_checkpoint 路径是否正确。")
        return

    # 3. 加载数据
    print("Loading WSI and Dataset...")
    try:
        # 使用自定义的 PILSlide 替代 openslide
        wsi = PILSlide(slide_path)
        dataset = Whole_Slide_Bag_FP(file_path=h5_path, wsi=wsi, target_patch_size=target_patch_size)
        loader = DataLoader(dataset, batch_size=batch_size, num_workers=0, collate_fn=collate_features, shuffle=False)
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return

    # 4. 提取特征循环
    print(f"Start processing {len(dataset)} patches...")
    features_list = []
    coords_list = []
    
    time_start = time.time()
    for count, (batch, coords) in enumerate(loader):
        with torch.no_grad():
            if count % 10 == 0:
                print(f'Batch {count}/{len(loader)} processed...')
            
            batch = batch.to(device)
            
            # BEPH/mmselfsup 模型推理
            # 注意：不同版本的 mmselfsup 接口可能略有不同，这里使用 backbone 提取
            if hasattr(model, 'module'):
                feat = model.module.extract_feat(batch, stage='backbone')[0] # [B, 768]
            else:
                feat = model.extract_feat(batch, stage='backbone')[0]
            
            # 对于 Vision Transformer，输出可能是 (B, N, C) 或 (B, C)
            # BEPH 通常使用 [CLS] token 或平均池化。mmselfsup 的 backbone 输出通常是一个 tuple
            # 如果是 (B, 197, 768)，我们需要取平均或取第一个。BEPH 论文通常取 [CLS] (index 0) 或 mean
            
            # 检查输出维度并做相应处理
            if len(feat.shape) == 3: 
                # (B, N_patches+1, Dim) -> 取 class token (index 0)
                feat = feat[:, 0, :] 
                
            features_list.append(feat.cpu().numpy())
            coords_list.append(coords)

    time_elapsed = time.time() - time_start
    print(f'\n提取完成！耗时: {time_elapsed:.2f} s')
    
    # 5. 整合并保存
    all_features = np.concatenate(features_list, axis=0)
    all_coords = np.concatenate(coords_list, axis=0)
    
    print(f'Features shape: {all_features.shape}')
    print(f'Coords shape: {all_coords.shape}')
    
    # 保存为 .h5
    with h5py.File(output_h5_name, 'w') as f:
        f.create_dataset('features', data=all_features)
        f.create_dataset('coords', data=all_coords)
    print(f'Saved H5 to: {output_h5_name}')
    
    # 保存为 .pt
    torch.save(torch.from_numpy(all_features), output_pt_name)
    print(f'Saved PT to: {output_pt_name}')

# 运行
run_feature_extraction()
```
